```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemma=WordNetLemmatizer()
ps = PorterStemmer()

def clean_Data(review_text):
    review_text = re.sub(r'http\S+', '', review_text)
    review_text = re.sub('[^a-zA-Z]',' ',review_text)
    #review_text = str(review_text).lower()
    review_text = word_tokenize(review_text)
    #for w in review_text : review_text = ps.stem(w)
    review_text = [item for item in review_text if item not in stop_words]
    review_text = [lemma.lemmatize(word=w,pos='v') for w in review_text]
    review_text = ' '.join(review_text)
    return review_text
```

```python
skill['CleanData'] = skill['RAW DATA'].apply(clean_Data)
```

# In Technology Skills, each skill's first letter of every word (except special symbol) is capital

```python
def is_capital(skill):
    flag = 0
    skills = []
    for kills in skill:
        flag = 0
        for word in kills.split():
            if word[0].islower():
                flag+=1
        if (flag>0):
            skills.append(0)
        else:
            skills.append(1)
    return skills
```

```python
req = is_capital(skill["RAW DATA"])
```

```python
req_skill = skill
req_skill["Flag"] = req
```
